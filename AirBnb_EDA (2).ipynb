#**1. Price & Value Insights**
Q: How does the average price differ across room types and neighborhoods?

```python
sns.barplot(x='room_type',y='price',data=df,errorbar=None,hue='neighbourhood_group',estimator='mean')
plt.title('Price vs Room Type Comparison')
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.show()
```

OBSERVATION:
This plot shows that **Hotel Rooms** has higher price in every **Neighbourhood Group** than other room types. And in Manhattan, Hotel Rooms has the highest prices as compared to Hotel Rooms in other cities.

#**2. Reviews and Popularity**
Q: Do cheaper listings tend to receive more reviews than expensive ones?

```python
sns.scatterplot(x='number_of_reviews_ltm',y='price',data=df,hue='neighbourhood_group')
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

OBSERVATION:
The plot reflects that property priced below 1000 get more reviews as compared to property costing Higher.
Which means that property below 1000 is affordable and hence more popular than the costlier one.

#**3. Host Activity**
Are hosts with multiple listings rated differently compared to single-listing hosts?

```python
df['host_type']=df.calculated_host_listings_count.apply(lambda x: "Mutiple" if x>1 else "Single")
```

```python
sns.barplot(x='host_type',y='rating',data=df,errorbar=None,estimator='mean')
plt.title('Price vs Host Type Comparison')
plt.xlabel('Host Type')
plt.ylabel('Price')
plt.show()
```

OBSERVATION:
The host having multiple listing has more average ratings, which means that the host with multiple properties provide better service and quality of living experience, that is the reason behing getting more ratings. On the other hand host with single listing does not manage property well as compared with the host with multiple listings.

#**4. Minimum Nights Policy**

Q: Do stricter minimum nights reduce the number of reviews?

```python
sns.scatterplot(x='minimum_nights',y='number_of_reviews_ltm',data=df,hue='neighbourhood_group')
plt.title('Minimum Nights vs Number of Reviews')
plt.xlabel('Minimum Nights')
plt.show()
```

OBSERVATION:
Lesser number of minimum nights got more number of reviews, which tells us that **lesser no. of minimum nights are preferred by the customers increasing flexiblity for them**.

#**6. Beds, Baths, and Price**

Q: How do the number of beds and baths affect price?

```python
sns.barplot(x='beds',y='price',data=df,errorbar=None)
plt.title('Price vs Bed Count')
plt.xlabel('Bed Count')
plt.ylabel('Price')
plt.show()

sns.barplot(x='baths',y='price',data=df,errorbar=None,estimator='mean')
plt.title('Price vs Bed Count')
plt.xlabel('Bath Count')
plt.ylabel('Price')
plt.show()
```

OBSERVATION:
1. When we look over the dependency of price over No. of beds, we see that price increases with the increase in bed count but, get decreased when the count of beds goes in range of 11 to 15. After that again continues the continuous growth.
2. Coming over the bathroom as a dependency we get to knpw it has static growth in price with increase in no.of bathrooms. Due to anomalies in dataset itself there are some lows in plot.

Colclusion: Price increases with the number of bedrooms and bathrooms.

#**7. Availability Patterns**

Q: Do listings available year-round (availability_365 = 365) have different review activity than seasonal listings?

```python
df['avl_status']=df.availability_365.apply(lambda x: "Year-Round" if x==365 else "Seasonal")
```

```python
sns.boxplot(x='avl_status',y='reviews_per_month',data=df)
plt.title('Availability Status vs Number of Reviews')
plt.xlabel('Availability Status')
```

OBSERVATION:
The properties which are available year round are tend to have lesser number of reviews per month, and for the seasonal availability the number of reviews per month are constantly growing, this means properties are occupies more in seasons also customer preffer to rent the the property occassionaly rather than renting it for the whole year.


#**📌Final Conclusion & Insights from EDA**

After analyzing the dataset across multiple dimensions such as price, reviews, host activity, availability, and property attributes, the following key insights have been derived:

#Room Type & Location Impact on Price

1.Hotel Rooms are the most expensive category across all neighbourhood groups.

2.Prices are highest in Manhattan, especially for Hotel Rooms, making it the most premium location compared to other areas.

#Price & Popularity Relationship

1.Properties priced below 1000 receive significantly more reviews, indicating that affordability drives popularity and customer engagement.

2.Higher-priced listings are less frequently reviewed, suggesting they cater to a more niche, less review-active audience.

#Host Activity & Service Quality

1.Hosts with multiple listings tend to have higher average ratings compared to single-property hosts.

2.This indicates that professional/experienced hosts may provide better quality service and consistency in guest experiences.

#Minimum Nights Policy & Guest Preference

1.Listings with lower minimum night requirements receive more reviews per month, showing that guests prefer flexibility and shorter stays.

2.Rigid or high minimum night policies limit customer interest.

#Beds, Baths, and Pricing Dependency

1.The price generally increases with the number of beds but shows a dip for properties with 11–15 beds, possibly due to anomalies or unique listings.

2.The price also increases steadily with the number of bathrooms, although occasional anomalies exist in the dataset.

Conclusion: Larger properties command higher prices, but extreme outliers may distort the trend.

#Availability & Seasonal Demand

1.Properties available year-round (365 days) receive fewer reviews per month compared to those with seasonal availability.

2.This suggests that customers prefer seasonal/occasional rentals rather than properties that are always open, likely aligning with tourism-driven demand cycles.

**For Hosts**: Offering flexible minimum night stays and keeping prices in the affordable range (<1000) can significantly boost demand and reviews.

**For New Entrants**: Investing in Manhattan Hotel Rooms may yield higher revenue potential, but competition is intense.

**For Customer Experience**: Scaling up to multiple listings appears beneficial, as professional hosting practices drive higher guest satisfaction.

**For Market Analysis**: The demand is seasonal and price-sensitive, so aligning availability and pricing with peak seasons can optimize occupancy and profitability.
